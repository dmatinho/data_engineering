#### Fill NaN values in order to convert columns with a dtype = float

```python
df6.loc[:,'DEP_TIME'].fillna(0, inplace=True)
df6.loc[:,'TAXI_OUT'].fillna(0, inplace=True)
df6.loc[:,'DEP_DELAY'].fillna(0, inplace=True)
df6.loc[:,'WHEELS_ON'].fillna(0, inplace=True)
df6.loc[:,'WHEELS_OFF'].fillna(0, inplace=True)
df6.loc[:,'TAXI_IN'].fillna(0, inplace=True)
df6.loc[:,'ACTUAL_ELAPSED_TIME'].fillna(0, inplace=True)
df6.loc[:,'AIR_TIME'].fillna(0, inplace=True)
df6.loc[:,'ARR_TIME'].fillna(0, inplace=True)
df6.loc[:,'ARR_DELAY'].fillna(0, inplace=True)
df6.loc[:,'CANCELLATION_CODE'].fillna('E', inplace=True)
df6.loc[:,'CARRIER_DELAY'].fillna(0, inplace=True)
df6.loc[:,'WEATHER_DELAY'].fillna(0, inplace=True)
df6.loc[:,'NAS_DELAY'].fillna(0, inplace=True)
df6.loc[:,'SECURITY_DELAY'].fillna(0, inplace=True)
df6.loc[:,'LATE_AIRCRAFT_DELAY'].fillna(0, inplace=True)
```

#### Convert colunms that refer to hours (integer values) into a time format

```python
df6.CRS_DEP_TIME.dtype
```

```python
def add_colon(some_num):
    if some_num ==2400:
       some_num = 0000 
    a = '{:02}:{:02}'.format(int(some_num)//100, int(some_num)%100)
    return a
```

```python
df6.loc[:,'CRS_DEP_TIME'] = df6.CRS_DEP_TIME.apply(add_colon)
df6.loc[:,'DEP_TIME'] = df6.DEP_TIME.apply(add_colon)
df6.loc[:,'WHEELS_OFF'] = df6.WHEELS_OFF.apply(add_colon)
df6.loc[:,'WHEELS_ON'] = df6.WHEELS_ON.apply(add_colon)
df6.loc[:,'CRS_ARR_TIME'] = df6.CRS_ARR_TIME.apply(add_colon)
df6.loc[:,'ARR_TIME'] = df6.ARR_TIME.apply(add_colon)
```
